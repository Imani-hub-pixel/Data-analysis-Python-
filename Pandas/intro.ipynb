```python
df.dropna(subset=['Events'], inplace=True)
```

```python
df_filled=df.fillna(0)
df_filled.isnull().sum()
```

```python
# Change the data type
df_filled.dtypes
```

```python
#Renaming columns
df=df_filled.rename(columns={'EST':'Date'})
df.head()
```

```python
df['Sea Level PressureInches']=df['Sea Level PressureIn'].astype(int)
```

```python
df.head(5)
```

```python
grouped_mean=df.groupby('Events')['Temperature'].mean()
print(grouped_mean)
```

```python
#Aggregate multiple functions
grouped_agg=df.groupby('Events')['Temperature'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
# Merging and joining DataFrames
df1=pd.DataFrame({'key':['A','B','C','D'],'value':[1,2,3,4]})
df2=pd.DataFrame({'key':['B','D','A','F'],'value':[5,6,7,8]})
```

```python
#Merging two dataframes
df3=pd.merge(df1,df2,on='key',how='inner')
print(df3)
```

```python
#Merging two dataframes
df3=pd.merge(df1,df2,on='key',how='outer')
print(df3)
```

```python
#Merging two dataframes
df3=pd.merge(df1,df2,on='key',how='left')
print(df3)
```

```python
#Merging two dataframes
df3=pd.merge(df1,df2,on='key',how='right')
print(df3)
```
