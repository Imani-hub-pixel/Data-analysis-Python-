```python
import pandas as pd
import matplotlib.pyplot as plt
sales_data_df=pd.read_csv('coffee_sales.csv')
sales_data_df.head()
```

```python
sales_data_df.isnull().sum()
```

```python
sales_data_df.info()
```

```python
#Plot money made by product
total_sales_by_product=sales_data_df.groupby('coffee_name')['money'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar', color='brown')
```

```python
#Plot sales Trend over time
sales_trend=sales_data_df.groupby('Monthsort')['money'].sum().reset_index()
print(sales_trend)
```

```python
sales_trend.plot(kind='line', color='green')
plt.title('Sales Trend over time')
plt.show()
```
