```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
#Imputation missing values
#Mean value imputation-works well when we we have normally distributed data
sns.displot(df['age'],kde=True)
```

```python
df['age_mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['age','age_mean']].head(10)
```

```python
sns.displot(df['age_mean'],kde=True)
```

```python
#Median value imputation-where we have outliers in the dataset
df['age_median']=df['age'].fillna(df['age'].median())
sns.displot(df['age_median'],kde=True)
```

```python
df['embarked'].unique()
```

```python
#Mode imputation-for categorical variables
df['embarked_filled']=df['embarked'].fillna(df['embarked'].mode()[0])
```
