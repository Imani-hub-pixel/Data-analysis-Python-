```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('jio_mart_items.csv')
df.head(5)
```

```python
df.shape
```

```python
df['category'].value_counts()
```

```python
from sklearn.utils import resample
target=10000
home_kitchen=df[df['category']=='Home & Kitchen']
groceries=df[df['category']=='Groceries']
fashion=df[df['category']=='Fashion']
electronics=df[df['category']=='Electronics']
Beauty=df[df['category']=='Beauty']
Jewellery=df[df['category']=='Jewellery']
```

```python
jewellery_upsampled=resample(
    Jewellery,
    replace=True,
    n_samples=target,
    random_state=42
)
balanced_df=pd.concat([
    home_kitchen,
    groceries,
    fashion,
    electronics,
    Beauty,
    jewellery_upsampled
    ])
```

```python
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```
