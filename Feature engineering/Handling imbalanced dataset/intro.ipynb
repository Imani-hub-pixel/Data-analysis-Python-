```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
```

```python
samples=1000
ratio=0.9
class_0=int(samples*ratio)
class_1=samples-class_0
```

```python
class_0,class_1
```

```python
class_0_data=pd.DataFrame({
    'feature_1': np.random.normal(loc=0.0,scale=1.0,size=class_0),
    'feature_2': np.random.normal(loc=0.0,scale=1.0,size=class_0),  
    'target': [0]*class_0

})
class_1_data=pd.DataFrame({
    'feature_1': np.random.normal(loc=2.0,scale=1.0,size=class_1),
    'feature_2': np.random.normal(loc=2.0,scale=1.0,size=class_1),  
    'target': [1]*class_1

})
```

```python
df=pd.concat([class_0_data,class_1_data],axis=0).sample(frac=1).reset_index(drop=True)
df.head()
```

```python
#Upsmapling 
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
df_upsampled['target'].value_counts()
```
