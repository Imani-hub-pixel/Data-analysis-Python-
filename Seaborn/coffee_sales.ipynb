```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coffee_sales = pd.read_csv('coffee_sales.csv')
coffee_sales.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='coffee_name',y='money',data=coffee_sales,estimator=sum)
plt.title('Total Money by Coffee Name')
plt.show()
```
