Feature engineering

```python
df.head()
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Minute']=df['Dep_Minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
import numpy as np
```

```python
df['Total_Stops']=df['Total_Stops'].map({
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4,
    np.nan:1

})
```

```python
df.head(2)
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df['Duration_hour']=df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['Duration_minute']=df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)
```

```python
df.info()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encode_df=encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df=pd.DataFrame(encode_df,columns=encoder.get_feature_names_out())
```

```python
new_df=pd.concat([df,encoded_df],axis=1)
```

```python
new_df.head(2)
```

```python
new_df.columns
```

```python
new_df.drop(['Airline','Source','Destination'],axis=1,inplace=True)
```
